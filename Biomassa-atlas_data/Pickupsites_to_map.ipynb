```python
import geopandas as gpd
import pandas as pd
import folium
import matplotlib.pyplot as plt
import os
```

```python
gdf = gpd.read_file("./nearest_pickup_sites.geojson")
```

```python
gdf
```

```python
gdf["Type"].unique()
```

```python
# Define a dictionary mapping types of data points to colors
color_dict = {
    'Horses and Ponies, Total Solid Manure from Storage': '#1f77b4',
    'Beef Cattle Liquid Manure from Animal Housing': '#ff7f0e',
    'Beef Cattle Total Solid Manure from Storage': '#2ca02c',
    'Broilers, Turkeys, and Other Poultry Total Solid Manure from Storage': '#d62728',
    'Sows and Piglets Liquid Manure from Animal Housing': '#9467bd',
    'Sows and Piglets Total Solid Manure from Storage': '#8c564b',
    'Side-stream: Grassland Residue': '#e377c2',
    'Side-stream: Straw': '#7f7f7f',
    'Side-stream: Dry Hayfields': '#17becf'
}
```

```python
## Initialize the Folium map
mymap = folium.Map(location=[gdf['geometry'].centroid.y.mean(), 
                             gdf['geometry'].centroid.x.mean()], 
                   zoom_start=10)

# Add markers with different colors based on the type of data point
for idx, row in gdf.iterrows():
    if row['Type'] in color_dict:
        color = color_dict[row['Type']]
    else:
        color = 'black'  # Default color if type not found in dictionary
    folium.CircleMarker(location=[row.geometry.y, row.geometry.x], radius=4, color=color, fill=True, fill_opacity = 1 ).add_to(mymap)
```
